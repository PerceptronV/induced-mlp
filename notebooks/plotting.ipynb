```python
import math
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
def plot_fn(min_x, max_x, fn, title=None, fname=None, x_label='x', y_label='y'):
    x = np.linspace(min_x, max_x, 100)
    y = fn(x)
    plt.plot(x, y)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    if title is not None:
        plt.title(title)
    plt.tight_layout()
    if fname is not None:
        plt.savefig(f'../media/graphs/{fname}.pdf')
    plt.show()
```

```python
f = 0.8
plot_fn(
    0, 1,
    lambda x: f * (np.sin(np.pi/2 * x)** 4),
    x_label='Training Progress',
    y_label="f'",
    title=f"Evolution of Triu Damping Factor f' (for f={f})",
    fname='dynamic-triu-damp'
)
```

```python
k = 0.3
plot_fn(
    0, 1,
    lambda x: 1 - (1-k) * (np.sin(np.pi/2 * x)** 4),
    x_label='Training Progress',
    y_label="k'",
    title=f"Evolution of TopK Factor k' (for k={k})",
    fname='dynamic-topk'
)
```
